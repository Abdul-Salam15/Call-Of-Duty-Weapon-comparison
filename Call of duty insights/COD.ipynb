```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Load the CSV file
df = pd.read_csv('CODM Weapons.csv')
#Input from user
name1 = input('Enter the Gun Name correctly -> ')
name2 = input('Enter the Gun Name correctly -> ')
GunNames = df['GUN NAME'].values
if name1 not in GunNames or name2 not in GunNames:
    print(f"One of the gun names was not correctly entered or not in the file")
else:
    # Filter the data for the specific weapons
    weapons_of_interest = df[df['GUN NAME'].isin([name1, name2])]
    
    # Extract the relevant columns
    weapon_names = weapons_of_interest['GUN NAME'].values
    damage_values = weapons_of_interest['DAMAGE'].values
    control_values = weapons_of_interest['CONTROL'].values
    mobility_values = weapons_of_interest['MOBILITY'].values
    range_values = weapons_of_interest['RANGE'].values
    accuracy_values = weapons_of_interest['ACCURACY'].values
    firerate_values = weapons_of_interest['FIRE RATE'].values
    # Create a DataFrame with the filtered data
    data = {
        'Weapon': weapon_names,
        'Damage': damage_values,
        'Fire Rate': firerate_values,
        'Mobility': mobility_values,
        'Accuracy': accuracy_values,
        'Range': range_values,
        'Control': control_values
                
    }
    df_filtered = pd.DataFrame(data)
    
    # Melt the DataFrame to long format for seaborn
    df_melted = df_filtered.melt(id_vars='Weapon', var_name='Attribute', value_name='Value')
    
    plt.figure(figsize=(10, 6))
    # Create the barplot
    sns.barplot(x='Attribute', y='Value', hue='Weapon', data=df_melted, palette='viridis')
    
    # Add titles and labels
    title = f'Comparison of {name1} and {name2} Attributes'
    plt.title(title)
    plt.xlabel('Attribute')
    plt.ylabel('Value')
    
    # Show the plot
    plt.show()
```
